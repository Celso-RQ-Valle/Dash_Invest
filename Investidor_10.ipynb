```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
import time
from selenium.common.exceptions import StaleElementReferenceException, TimeoutException
import numpy as np
from datetime import datetime

# Configure o webdriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

url = 'https://investidor10.com.br/acoes/busca-avancada/'
driver.get(url)
```

```python
# Lendo o primeiro elemento
try:
    # Aguardando até que o elemento da tabela esteja presente na página
    element_present = EC.presence_of_element_located((By.XPATH, '//*[@id="table-advanced-search"]'))
    WebDriverWait(driver, 20).until(element_present)
except Exception as e:
    print(f"Erro ao carregar a tabela: {e}")

try:
    # Encontre o elemento da tabela usando o XPath
    table_element = driver.find_element(By.XPATH, '//*[@id="table-advanced-search"]')

    # Extraia o HTML da tabela
    table_html = table_element.get_attribute('outerHTML')

    # Parsear o HTML usando BeautifulSoup
    soup = BeautifulSoup(table_html, 'html.parser')

    # Encontrar a tabela
    table = soup.find('table')

    # Converter a tabela para um DataFrame do pandas
    df = pd.read_html(str(table))[0]

    print(df)
except Exception as e:
    print(f"Erro ao processar a tabela: {e}")
```

### Formatando Base Investidor 10

```python
guarda = df.copy()
```

```python
df = guarda.copy()
def replace_dash(value):
    return '' if value == '-' else value

col = df.columns[2:]
for k in range(len(col)):
    try:
        # Tenta limpar e converter valores com formatação
        df[col[k]] = df[col[k]].apply(replace_dash).replace(r'[,.]', '', regex=True)\
                               .replace(r'^\s*$', np.nan, regex=True).astype(float) / 100
    except ValueError:
        # Caso a conversão falhe, trata a coluna conforme seu tipo
        if df[col[k]].dtype == 'int64':
            df[col[k]] = df[col[k]].astype(float) / 100
        elif df[col[k]].dtype == 'O':
            df[col[k]] = df[col[k]].apply(replace_dash).replace('-', '').replace(r'^\s*$', np.nan, regex=True)
            df[col[k]] = df[col[k]].str.replace(r'[.%mil]', '', regex=True)
            #convertendo , forma decimal br para . forma decimal eua
            df[col[k]] = df[col[k]].replace(r'[,]', '.', regex=True).astype(float) / 100
            df[col[k]] = df[col[k]].round(4)

# Reinicie os índices do DataFrame após a alteração
df.reset_index(drop=True, inplace=True)
```
